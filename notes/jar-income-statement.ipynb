```python
year = 2021
df = pd.read_csv(f'https://www.registrucentras.lt/aduomenys/?byla=JAR_FA_RODIKLIAI_PLNA_{year}.csv')
df.drop(columns=['stat_pav', 'template_name', 'standard_name', 'template_id', 'standard_id', 'formavimo_data', 'form_pav'], inplace=True)
```

```python
df.laikotarpis_nuo = pd.to_datetime(df.laikotarpis_nuo)
df.laikotarpis_iki = pd.to_datetime(df.laikotarpis_iki)
df.reg_date = pd.to_datetime(df.reg_date)
```

```python
uab = df[df['obj_kodas'] == 125483974]
assert len(uab) == 1
uab
```

```python
no_revenue = df[df.pardavimo_pajamos == 0 | df.pardavimo_pajamos.empty]
no_revenue_percent = len(no_revenue) / len(df)
print(f'Percentage of businesses that had no revenue in {year}: {no_revenue_percent:.2%}')
no_revenue
```

```python
outdated = df[df.laikotarpis_iki < pd.Timestamp(f'{year - 1}')]
outdated_percent = len(outdated) / len(df)
print(f'Percentage of businesses that have outdated have provided outdated (?) data in {year}: {outdated_percent:.2%}')
outdated
```
